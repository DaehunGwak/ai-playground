# 6. 지금까지 배운 내용 총 정리 (feat. 프롬프트 꿀팁 살짝)

- 이번엔 Country > Food > Recipe 이렇게 만들어보자 (일단 혼자 레츠고)
- 줄바꿈도 토큰... ㅠ
- `Please` 를 쓰면 더 잘해준다는 논문이 있음. 그럼 한글로는 존댓말이지 않을까?
- LLM 을 역할을 나눠서 간결하게 작성하는게 좋음

```python
from langchain_ollama import ChatOllama
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate

llm = ChatOllama(model ="deepseek-r1:8b")
output_parser = StrOutputParser()
```

```python
country_to_food_template = PromptTemplate(
    template="Return the food of {country}. Please, return only the food name.",
    input_variables=["country"],
)

country_to_food_chain = country_to_food_template | llm | output_parser

country_to_food_chain.invoke({'country': '오스트레일리아'})
```

```python
food_to_recipe_template = ChatPromptTemplate.from_messages([
    ("system", """Provide the recipe of the food that the user wants. 
    Please return only the recipe as a numbered list."""),
    ("human", "Can you give me the recipe for making {food}?")
])

food_to_recipe_chain = food_to_recipe_template | llm | output_parser

food_to_recipe_chain.invoke({'food': 'Vegemite'})
```

```python
country_to_recipe_chain = {'food': country_to_food_chain} | food_to_recipe_chain

country_to_recipe_chain.invoke({'country': '프랑스'})
```
