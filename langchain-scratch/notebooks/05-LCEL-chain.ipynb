# 5. LCEL을 활용한 랭"체인" 생성하는 방법

랭체인 모든게 `Runnable` 을 상속하는 구조

```python
from langchain_ollama import ChatOllama

llm = ChatOllama(model ="deepseek-r1:8b")
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate(
    template="what is the capital of {country}? Return only the city name.", # prompt 도 city 만 받도록 수정
    input_variables=["country"],
)

prompt = prompt_template.invoke({'country': '오스트레일리아'})
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
```

```python
capital_str_chain = prompt_template | llm | output_parser
```

```python
capital_str_chain.invoke({'country': '오스트레일리아'})
```

```python
country_prompt = PromptTemplate(
    template="""
    Guess the name of the country based on the following information.
    {information}

    Return the name of the country only.
    """,
    input_variables=["information"],
)

country_chain = country_prompt | llm | output_parser
country_chain.invoke({'information': 'This country is very famouse for its wine.'})
```

```python
final_chain = {"country": country_chain} | capital_str_chain

final_chain.invoke({"information": "This country is very famouse for its wine."})
```

```python
from langchain_core.runnables import RunnablePassthrough

final_chain = {'information': RunnablePassthrough()} | country_chain | capital_str_chain
```

```python
final_chain.invoke("This country is very famouse for its wine.")
```
