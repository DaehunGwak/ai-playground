- ollama 활용 예정

```shell
brew install ollama
```

````python
import os

from langchain_ollama import ChatOllama

llm = ChatOllama(model ="deepseek-r1:8b")
````

````python
llm.invoke("넌 지금 어디에 있니?")
````

다시 설치
```shell
uv add langchain-anthropic # https://docs.langchain.com/oss/python/integrations/chat/anthropic
uv add python-dotenv
```


.env 파일
```shell
# .env 파일에서
ANTHROPIC_API_KEY=<your_api_key>
```

````python
from dotenv import load_dotenv

load_dotenv()
````

````python
from langchain_anthropic import ChatAnthropic

model = ChatAnthropic(
    model="claude-haiku-4-5-20251001",
    temperature=0,
    max_tokens=1024,
    timeout=None,
    max_retries=2,
    # other params...
)
````

````python
message = [
    ("system", ""),
    ("human", "안녕 넌 어디에 있니?")
]
ai_msg = model.invoke(message)
ai_msg
````
