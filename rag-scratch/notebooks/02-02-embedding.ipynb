```python
from langchain_ollama import OllamaEmbeddings

gemma_embeddings = OllamaEmbeddings(model="embeddinggemma:300m")
```

```python
gemma_embeddings
```

```python
ko_king_vector = gemma_embeddings.embed_query(text='왕')
en_king_vector = gemma_embeddings.embed_query(text='King')

len(ko_king_vector), len(en_king_vector)
```

```python
import numpy as np

def cosine_similarity(vec1, vec2):
    """두 벡터 간의 코사인 유사도 계산"""
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)

cosine_similarity(ko_king_vector, en_king_vector)
```

```python
cosine_similarity(
    gemma_embeddings.embed_query(text='DEAN'),
    gemma_embeddings.embed_query(text='딘')
)
```

```python
cosine_similarity(
    gemma_embeddings.embed_query(text='DEAN'),
    gemma_embeddings.embed_query(text='지코')
)
```
