```python
import dotenv, os
from supabase import create_client

dotenv.load_dotenv("../.env.local", override=True)

supabase_client = create_client(
    supabase_url=os.getenv("SUPABASE_URL"),
    supabase_key=os.getenv("SUPABASE_SECRET_KEY"),
)
```

```python
count_result = supabase_client.schema('public').table('tracks') \
    .select('*', count='exact', head=True) \
    .execute()
count_result.count
```

```python
result = supabase_client.schema('public').table('tracks') \
    .select('*') \
    .range(0, 10) \
    .execute()
result.data[0]['audioUrl'] = 'https://sample.audio.url.nontivi'
result.data[0]
```

```python
# Pagination over all rows in the 'tracks' table, 10 at a time
page_size = 10
offset = 0
all_tracks = []

while True:
    page_result = supabase_client.schema('public').table('tracks') \
        .select('*') \
        .range(offset, offset + page_size - 1) \
        .execute()
    data = page_result.data
    if not data:
        break
    all_tracks.extend(data)
    if len(data) < page_size:
        break
    offset += page_size

# Optional: print how many tracks were loaded
print(f"Total tracks loaded: {len(all_tracks)}")
```
