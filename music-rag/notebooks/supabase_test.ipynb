```python
import dotenv, os
from supabase import create_client

dotenv.load_dotenv("../.env.local", override=True)

supabase_client = create_client(
    supabase_url=os.getenv("SUPABASE_URL"),
    supabase_key=os.getenv("SUPABASE_SECRET_KEY"),
)
```

```python
supabase_client.schema('public').table('tracks').select('*').execute()
```

```python
supabase_client.schema('public').table('tracks').insert({
    'title': 'Sample Track',
    'artists': 'Test Artist',
    'tags': 'pop, electronic',
    'audioUrl': '',
    'cpTrackId': 'uuid-sample',
    'imageUrl': ''
}).execute()
```

```python
# Supabase Storage에 mp3 파일 업로드
# 주의: Supabase 대시보드에서 먼저 'musics' 버킷을 생성해야 합니다

file_path = "../static_files/musics/Untitled by Nontivi.mp3"
bucket_name = "musics"  # Supabase에서 생성한 버킷 이름
storage_path = "Untitled by Nontivi.mp3"  # Storage 내 저장될 경로/파일명

with open(file_path, "rb") as f:
    file_data = f.read()

# 파일 업로드
response = supabase_client.storage.from_(bucket_name).upload(
    path=storage_path,
    file=file_data,
    file_options={"content-type": "audio/mpeg"}
)

print(response)
```
